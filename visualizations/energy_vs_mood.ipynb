```python
import pandas as pd
import altair as alt

df = pd.read_csv('../data/visualizations/energy_valence_scatter.csv')

input_radio = alt.binding_radio(options = [None, 'Acoustic/Folk', 'Country', 'Electronic/Dance', 
                                           'Hip-Hop/Rap', 'Jazz/Blues', 'Latin',
                                           'Other', 'Pop', 'R&B/Soul', 'Rock/Alternative'],
                                    labels = ['All', 'Acoustic/Folk', 'Country', 'Electronic/Dance', 
                                           'Hip-Hop/Rap', 'Jazz/Blues', 'Latin',
                                           'Other', 'Pop', 'R&B/Soul', 'Rock/Alternative'],
                                     name = "Genre: ")

selection = alt.selection_point(fields = ['macro_genre'], bind = input_radio)

chart = (alt.Chart(df).mark_circle().encode(
    x = alt.X('energy:Q'),
    y = alt.Y('valence:Q'),
    color = alt.condition(selection, alt.Color('macro_genre:N'), alt.value('grey')),
    tooltip=['track_name', 'artist_spotify', 'macro_genre', 'energy', 'valence']
).properties(
    title = "Energy vs Mood of Songs by Genre",
    width=600,
    height=400
).add_params(
    selection
).transform_filter(
    selection
))
chart

chart.save("../visualizations/energyvsmood.html")


```
